```python
import cv2

# 눈 검출기 Haar Cascade XML 파일 경로
eye_cascade_path = cv2.data.haarcascades + "haarcascade_eye.xml"

# Haar Cascade 눈 검출기 객체 생성
eye_cascade = cv2.CascadeClassifier(eye_cascade_path)

cap = cv2.VideoCapture(0)  # 카메라 캡처 객체 생성

# 이전 눈 중심의 초기 값
last_eye_center = None

# 눈동자 움직임 횟수 초기화
move_count = 0

while True:
    res, frame = cap.read()  # 카메라 프레임 읽기
    frame = cv2.flip(frame,1)

    if not res:
        break

    # 그레이스케일 변환
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 히스토그램 평활화
    gray = cv2.equalizeHist(gray)

    # 눈 검출
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in eyes:
        # 눈의 중심 계산
        eye_center = (x + w // 2, y + h // 2)

        # 눈동자를 감싸는 사각형 그리기
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # 눈동자 중심 표시
        cv2.circle(frame, eye_center, 2, (0, 255, 0), -1)

        # 이전 눈 중심과 현재 눈 중심의 X 좌표 차이 계산
        if last_eye_center is not None:
            movement = abs(eye_center[0] - last_eye_center[0])

            # 눈동자 이동 감지
            if movement > 10:  # 예시로 임계값을 10으로 설정
                move_count += 1
                print("Eye movement detected!")

        # 현재 눈 중심을 이전 눈 중심으로 업데이트
        last_eye_center = eye_center

    cv2.imshow("Frame", frame)

    key = cv2.waitKey(1) & 0xFF
    if key == 27:  # ESC 키를 누르면 종료
        print(move_count)
        break

# 리소스 해제
cap.release()
cv2.destroyAllWindows()
```
