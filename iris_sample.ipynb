```python
import pandas as pd #데이터 시각화
from sklearn.datasets import load_iris # sklearn에서 iris 샘플 데이터 불러오기
from sklearn.model_selection import train_test_split # 데이터 쪼개기
from sklearn.naive_bayes import GaussianNB # 가우시안 나이브 베이즈

dataset = load_iris()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names) # 데이터 DataFrame으로 저장

df['target'] = dataset.target # 타깃 레이블을 데이터에 저장
df.target = df.target.map({0:"setosa", 1:"versicolor", 2:"virginica"}) 
# target을 문자열로 나타내기

df.head()

setosa_df = df[df.target == "setosa"]
versicolor_df = df[df.target == "versicolor"]
virginica_df = df[df.target == "virginica"]

ax = setosa_df['sepal length (cm)'].plot(kind='hist')
#kde = 커널 밀도 추정 (히스토그램 같은 분포를 부드럽게 곡선화 시켜서 그려주는 것)
setosa_df['sepal length (cm)'].plot(kind='kde', ax=ax, secondary_y=True,
                                    title="setosa sepal length (cm) distribution",
                                    figsize=(8,4))

```
